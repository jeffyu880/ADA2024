```python
import pandas as pd

df = pd.read_csv('./data/interactive_data.csv')

print(df.to_string()) 
```

```python
df_cleaned = df[~df.isin(['None selected']).any(axis=1)]
# print(df_cleaned)

Suicide_rows = df_cleaned[df_cleaned['Intent'] == 'Suicide']
numDeaths = df_cleaned['Deaths'].sum()
numSuicideDeaths = Suicide_rows['Deaths'].sum()
# print(Suicide_rows.shape[0])

print('Percentage of suicides: ', numSuicideDeaths/numDeaths)
```

```python
numMaleSuicides = Suicide_rows[Suicide_rows['Gender'] == 'Male']['Deaths'].sum()
print("Percentage of male suicides: ", numMaleSuicides/numSuicideDeaths)
```

```python
numHomicides = df_cleaned[df_cleaned['Intent'] == 'Homicide']['Deaths'].sum()
print('Percentage of Homicides: ', numHomicides/numDeaths)
```

```python
numMalesHomicideinAgeGrp = df_cleaned[(df_cleaned['Gender'] == 'Male') &
                                        (df_cleaned['Intent'] == 'Homicide') &
                                        (df_cleaned['Age'] == '15 - 34')]['Deaths'].sum()

numBlackHomicides = df_cleaned[(df_cleaned['Gender'] == 'Male') &
                        (df_cleaned['Intent'] == 'Homicide') &
                        (df_cleaned['Age'] == '15 - 34') &
                        (df_cleaned['Race'] == 'Black')]['Deaths'].sum()

print('Percentage of Black homicides: ', numBlackHomicides/numMalesHomicideinAgeGrp)
```

```python
numWomanHomicides = df_cleaned[(df_cleaned['Gender'] == 'Female') &
                                (df_cleaned['Intent'] == 'Homicide')]['Deaths'].sum()

print('Percentage of woman homicides: ', numWomanHomicides/numHomicides)
```
